```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Classdata_df = pd.read_csv("CSCI_Fall_2024_survey.csv")
Classdata_df["Age"] = Classdata_df["How old are you?"].replace({
    17: 17, 
    18: 18, 
    19: 19, 
    20: 20, 
    21: "Unknown"
})
agg_Age = Classdata_df["Age"].value_counts()
colors = ['pink', 'blue', 'yellow', 'purple', 'red']
agg_Age.plot(kind="pie", colors=colors)
plt.title('Age Distribution')
plt.ylabel('')
plt.show()

#This is a pie graph that shows the distribution of age in our class.

```

```python
#Distribution of Majors,HIST
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
Classdata_df = pd.read_csv("CSCI_Fall_2024_survey.csv")
Classdata_df["Major"] = Classdata_df["What is your major?"].replace({
    "Economics": "ECON",
    "computer science": "Computer Science",
    "Data Science": "Data Science",
    "Biology": "Biology",
    "CS/Business": "CS/Business",
    "cs/finance": "CS/Finance"
})

Major_count = Classdata_df["Major"].value_counts()
Major_count.plot(kind="bar")
plt.xlabel('Major')
plt.ylabel('Count')
plt.title('Histogram of Major')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid()
plt.show()

```

```python
# This should return as a pie chart for how many people like food in country categories. 

# Standardize food preferences
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
classdata_df = pd.read_csv("CSCI_Fall_2024_survey.csv")

# Standardize food preferences
classdata_df["Food_df"] = np.where(classdata_df["What is your favorite type of food?"] == "Japanese", "Japan",
                    np.where(classdata_df["What is your favorite type of food?"] == "korean", "Korea",
                    np.where(classdata_df["What is your favorite type of food?"] == "Italian", "Italy",
                    np.where(classdata_df["What is your favorite type of food?"] == "American", "USA",
                    np.where(classdata_df["What is your favorite type of food?"] == "Mexican", "Mexico",
                    np.where(classdata_df["What is your favorite type of food?"] == "Indian", "India",
                    np.where(classdata_df["What is your favorite type of food?"] == "Chinese", "China", "all of the above")))))))
agg_food = classdata_df["Food_df"].value_counts()

colors = ['pink', 'blue', 'green', 'orange', 'yellow', 'purple', 'red']  # Adjust colors as needed
agg_food.plot(kind="pie", y = "Food", colors=colors)

# Customize plot
plt.title('Food Preference')
plt.ylabel('')  # Remove y-label for better appearance
plt.show()
```

```python

#How many pets and of which kind do you have?, HISY 
#This is supposed to be a histogram showing how many pets people have.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
classdata_df = pd.read_csv("CSCI_Fall_2024_survey.csv")
# Standardize pets
Classdata_df["Pets"] = np.where(Classdata_df["How many pets and of which kind do you have?"] == "0", "zero pets",
                     np.where(Classdata_df["How many pets and of which kind do you have?"] == "1", "one pet",
                     np.where(Classdata_df["How many pets and of which kind do you have?"] == "I have two dogs and they are both at home", "two pets",
                     np.where(Classdata_df["How many pets and of which kind do you have?"] == "2 cats, one dog", "three pets",
                     np.where(Classdata_df["How many pets and of which kind do you have?"] == "6 Dogs (Over two households)", "six pets",
                     np.where(Classdata_df["How many pets and of which kind do you have?"] == "had 8 dogs, not anymore", "eight pets", "no pets"))))))
plt.plot(Classdata_df["Pets"], Classdata_df["Major"],'.')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
classdata_df = pd.read_csv("CSCI_Fall_2024_survey.csv")

#standardize Birthplace
# Create a mapping dictionary
birth_place_map = {
    "I was born in Chandigarh, India!": "India",
    "Kazakhstah": "Kazakhstan",
    "Cameroon": "Cameroon",
    "People's Republic of China": "China",
    "Georgia": "Georgia",
    "India": "India",
    "Frankfurt, Germany": "Germany",
    "Mexico": "Mexico",
    "Wisconsin": "Wisconsin",
    "New Jersey": "New Jersey",
    "Oklahoma": "Oklahoma",
    "Kansas, Overland Park": "Kansas",
    "Missouri": "Missouri",
    "st. louis, mo": "Missouri",
    "St. Louis MO": "Missouri",
    "Missouri, USA": "Missouri",
    "Saint Louis, Missouri": "Missouri"
}

# Map the values and create the new column
Classdata_df["Birth"] = Classdata_df["What state/country were you born in?"].map(birth_place_map).fillna("Unknown")

Birth_count= Classdata_df["Birth"].value_counts()

Birth_count.plot(kind="bar")
plt.xlabel('Birth')
plt.ylabel('Count')
plt.title('Histogram of Birth')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid()

#This is a histogram about the birthplaces of people outside the US and states inside the US
```
